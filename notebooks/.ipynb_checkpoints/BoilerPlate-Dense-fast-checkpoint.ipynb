```python
import itertools
import gc
import timeit
from bib import *
import numpy as np
import pandas as pd
from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import backend as K 
from keras import regularizers
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
xtrains,ytrains,xvals,yvals,xtest,ytest=split(('CHL 1','THERM 1','SSH','SR','WS'),
                                               ['CHL '+ str(i) for i in range(2,19)])
ytrains,yvals,ytest=applyToy(lambda x:np.log(x)/np.log(10),ytrains,yvals,ytest)
```

```python
def generate(arch,act):
    layers=[]
    for index,size in enumerate(arch):
        if index ==0:
            layers.append(Dense(size, input_shape=(5,),activation=act[index]))

        else:
            layers.append(Dense(size,activation=act[index]))
    layers.append(Dense(17,activation='linear',kernel_regularizer=regularizers.l2(0.01)))
    model = Sequential(layers)
    model.compile(optimizer='adam',
              loss='mse')
    return model
callbacks = [EarlyStopping(monitor='val_loss', patience=5)]
```

```python
def validate(arch,act,xtrains,ytrains,xvals,yvals):
    fold_evaluation=[]
    for xt,yt,xv,yv in zip(xtrains,ytrains,xvals,yvals):
        model=generate(arch,act)
        
#         mean=xt.mean()
#         std=xt.std()
        
#         xt=(xt-mean)/std
#         xv=(xv-mean)/std
        
        
        
        model.fit(xt.values,yt.values,
                  epochs=10000,
                  callbacks=callbacks,
                  verbose=0,
                  batch_size=73*10, 
                  validation_data=(xv.values,yv.values))
        
        yp = model.predict(xv.values)
        
        #check error in original space
        yp=np.exp(yp*np.log(10))
        yv=np.exp(yv*np.log(10))
        fold_evaluation.append(evaluate(yp,yv))
        
    return np.mean(fold_evaluation),np.std(fold_evaluation )
```

```python
#normalise
for i in range(len(xvals)):
    mean=xtrains[i].mean()
    std=xtrains[i].std()
    xtrains[i]=(xtrains[i]-mean)/std
    xvals[i]=(xvals[i]-mean)/std
```

```python




layersN=[1,2,3,4,5,6,7]
neuronsN=[10,50,100,200,500]
# acts=['relu','sigmoid','softplus']
results=[]
for N in layersN :
    for arch in itertools.product(neuronsN,repeat= N):
            print('ARCH :'+ str(arch))
            start = timeit.default_timer()
            error,std=validate(arch,['softplus']*N,xtrains,ytrains,xvals,yvals)
            stop = timeit.default_timer()
            time=stop-start
            print('\t error: %.5f +- %.5f time %.3f' % (error,std,time))
            results.append([arch,error,std,time])
            gc.collect()
```

```python




layersN=[1,2,3,4,5,6,7]
neuronsN=[10,50,100,200,500]
# acts=['relu','sigmoid','softplus']
results=[]
resume=False
for N in layersN :
    for arch in itertools.product(neuronsN,repeat= N):
        if arch == (100, 10, 200):
            resume =True
        if resume:
            print('ARCH :'+ str(arch))
            start = timeit.default_timer()
            error,std=validate(arch,['softplus']*N,xtrains,ytrains,xvals,yvals)
            stop = timeit.default_timer()
            time=stop-start
            print('\t error: %.5f +- %.5f time %.3f' % (error,std,time))
            results.append([arch,error,std,time])
            gc.collect()
            K.clear_session()
```
